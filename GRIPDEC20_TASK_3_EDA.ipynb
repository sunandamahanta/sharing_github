## STEP 3 : Exploratory Data Analysis


### 1. UNIVARIATE ANALYSIS

#### CATEGORICAL COLUMNS :

```python
#Ship Mode
sns.countplot(x=data['Ship Mode'])
plt.xticks(rotation=60)
plt.show()
```

##### INSIGHT : Most of the shipments are done under 'Standard Class' , on the other hand 'Same Day' shipments are very less.

```python
#Segment
sns.countplot(x=data['Segment'])
plt.xticks(rotation=60)
plt.show()
```

##### INSIGHT - Consumers (who purchases goods and services for personal use) have purchased the most that the corporate people and the people who purchase for their home office.

The data is about only one country USA. So we will not visualize that column and proceed with 'City' & 'State' columns.

```python
#State
state = data['State'].value_counts()
plt.figure(figsize=(15,16))
sns.countplot(x=data['State'])
plt.xticks(rotation=90)
plt.show()
```

##### INSIGHT : Most of the shipments are done to the country California, followed by New York, Texas.

We have 531 cities in our data set. It won't be possible to visualize such a large data in bar graph. So we will visualize the 'City' column using a line plot.

```python
#Grouping Cities as per their respective States and counting number of orders in each city
order = data.groupby(['State','City']).size().reset_index(name='No of orders')
```

```python
#Sorting to check which city/state has the highest number of orders.
order.sort_values(by='No of orders',ascending=False).head()
```

```python
#City 
city = data['City'].value_counts()
city.plot(kind='line',figsize=(8,5))
plt.title("Orders from the Cities")
plt.ylabel('Count')
plt.xlabel('City')
plt.show()
```

##### INSIGHT : From the above line plot, it can be seen that New York City has the highest purchase/sellings among all other cities.

```python
#Cities with minimum order
order[order['No of orders']==order['No of orders'].min()]
```

```python
#Number of cities with least number of orders
order[order['No of orders']==order['No of orders'].min()].shape[0]
```

##### Since there are 78 cities with least number orders, we can analyze the problem and try to solve them in order to increase sales in those areas.

```python
#Grouping City, State according the regions
cnt = data.groupby(['City', 'State', 'Region']).size().reset_index(name='Count')
```

```python
cnt.sort_values(by='Count',ascending=False).head()
```

```python
#Region
sns.countplot(x=data['Region'])
plt.xticks(rotation=60)
plt.show()
```

##### INSIGHT : West & East regions have high sales. We need to keep in mind that the products don't go out of stock.

```python
#Sales vs Profit
sales_profit = data[['Sales','Profit']].sort_values('Profit',ascending=False).head(30)
plt1 = sns.scatterplot(x='Sales',y='Profit',data=sales_profit)
plt.xticks(rotation=90)
plt.show()
```

```python
#Region vs category
pd.crosstab( data['Region'],data['Category']).plot(kind='bar', figsize=(10,7))
plt.xticks(rotation=0)
plt.plot()
```

For every region maximum sell is in office category.

```python
# city/state/region wise profit
profit_city = data.groupby('City').sum()['Profit']
```

```python
fig, axes = plt.subplots(1,2, figsize=(15,5))
plt.xticks(rotation=45)
profit_city.sort_values(ascending=False).head().plot(kind='bar', ax=axes[0])
plt.xticks(rotation=60)
plt.title('Maximum Profit')

profit_city.sort_values(ascending=True).head().plot(kind='bar', ax=axes[1])
plt.xticks(rotation=60)
plt.title('Minimum Profit')
plt.show()
```

##### NYC gains the most profit and Philadelphia is facing the most loss.

```python
# state wise profit distribution
state_profit = data.groupby('State').sum()['Profit']
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=state_profit.index, y=state_profit.values)
plt.xticks(rotation=90)

plt.show()
```

```python
# region wise profit distribution
region_profit = data.groupby('Region').sum()['Profit']
```

```python
plt.pie(region_profit, labels=region_profit.index, autopct='%.2f', explode=(0,0,0,0.1), shadow=True)
plt.show()
```

##### INSIGHT: East and West regions are making more profit.
